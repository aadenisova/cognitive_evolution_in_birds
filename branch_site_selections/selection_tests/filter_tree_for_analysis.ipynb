```python
from ete3 import Tree
import pandas as pd
```

```python
PATH_TO_PROJECT="/lustre/fs5/jarv_lab/scratch/adenisova/Inno_2025"

tree_path = f"{PATH_TO_PROJECT}/initial_data/trees_labeled/pruned_tree_TOGA_inno_marked.newick"

df = pd.read_csv(f"{PATH_TO_PROJECT}/tmp/alignment_qc_summary.tsv", sep="\t")
df_filtered = df[
    (df["n_species"] >= 300) &
    (df["gap_columns_frac"] <= 0.13) &
    (df["gap50_columns_frac"] <= 0.19) &
    (df["number_of_foreground_species"] >= 23) 
]
df_filtered

foreground_df = pd.read_csv(f"{PATH_TO_PROJECT}/initial_data/foreground_species.tsv", sep = "\s+")
foreground_df = foreground_df[foreground_df["Inno"] == "high"]
foreground_df = set(foreground_df["Scientific_Name"])
foreground_df
```

```python
",".join([i+"{TEST}" for i in foreground_df])
```

```python
for idx, row in df_filtered.iterrows():
    t = Tree(tree_path)

    sp_foreground_list = [i+"{TEST}" for i in foreground_df]
    background_sp = list((set(row["all_species"].split(",")) - foreground_df))
    all_list = sp_foreground_list + background_sp
    t.prune(all_list)
    print(t)

    break

```
