```python
import pandas as pd
from statsmodels.stats.multitest import multipletests

PATH = "/lustre/fs5/jarv_lab/scratch/adenisova/Inno_2025/genome_data/ncbi_dataset/data"
df = pd.read_csv("/lustre/fs5/jarv_lab/scratch/adenisova/Inno_2025/genome_data/ncbi_dataset/data/busted_e_vs_busted_table.csv")
df
```

```python
import os

df_sign = df[
    (df['p-value'] < 0.05) &
    (df['p-value_e'] < 0.05)
]

n_in_file = round(df_sign["gene"].shape[0]/100)
os.makedirs(f"{PATH}/batches_absrel", exist_ok=True)

n = 1
iter = 0 

file = open(f"{PATH}/batches_absrel/batch_{iter}.txt", "w")
for og in df_sign["gene"]:
    print(og)
    print(iter)
    print(n)
    print(f"{PATH}/batches_absrel/batch_{iter}.txt")
    
    if n >= n_in_file:
        n = 0
        iter += 1
        file.close()
        file = open(f"{PATH}/batches_absrel/batch_{iter}.txt", "w")

    file.write(f"alignments/{og}.macse.clean.phy\n")
    n += 1

file.close()

    
```
